```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from langchain_core.messages import HumanMessage
```

```python
repo_id = "meta-llama/Llama-3.2-1B-Instruct"
llm=HuggingFaceEndpoint(repo_id=repo_id, temperature=0.7, task="conversational", huggingfacehub_api_token=os.getenv("HF_TOKEN"))
```

```python
# 2. WRAP IT in ChatHuggingFace to fix the Novita error
chat_model = ChatHuggingFace(llm=llm)

# 3. Use it by passing a list of messages
messages = [HumanMessage(content="Hello! How are you?")]
chat_model.invoke(messages)

# print(response.content)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
template="""
Question:{question}
Answer: lets think step by step
"""
```

```python
prompt=PromptTemplate(template=template, input_variables=['question'])
```

```python
# 2. WRAP IT in ChatHuggingFace (This is the crucial step)
chat_model = ChatHuggingFace(llm=llm)

# 3. Create your prompt
prompt = PromptTemplate.from_template("Answer this question: {question}")

# 4. Use the wrapped chat_model in your chain
llm_chain = prompt | chat_model

# 5. Invoke the chain
response = llm_chain.invoke({"question": "who won cricket world cup 2011"})

print(response.content)
```
